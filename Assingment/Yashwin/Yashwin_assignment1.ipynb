## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import pandas as pd
import datetime

tdate= datetime.datetime.strptime("01-01-2023", "%d-%m-%Y")
date=pd.date_range(tdate,41)
print(date.strftime("%d-%m-%Y"))
```
